```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/AI test/loan_payment analysis/Loan payments data_원본.csv")

df.head()
```

```python
df.drop(['effective_date', 'due_date' , 'paid_off_time', 'past_due_days'] ,axis = 1, inplace = True)
df.head()
```

```python
df["education"]= df["education"].replace(['High School or Below', 'college', 'Bechalor', 'Master or Above'],["0","1","2","3"])
df
```

```python
df["Gender"]= df["Gender"].replace(['female', 'male'],["0","1"])
df
```

```python
df["loan_status"]= df["loan_status"].replace(['PAIDOFF', 'COLLECTION', 'COLLECTION_PAIDOFF'],["0","1","2"])
df
```

```python
df['Loan_ID'] =df['Loan_ID'].str[3:]
df
```

```python
df.to_csv('Loan payments data_PreProcessed.csv')
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/Loan payments data_PreProcessed.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[1:,1:].values
y = df.iloc[1:,2].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
print(X_train)
print(y_train)
```
